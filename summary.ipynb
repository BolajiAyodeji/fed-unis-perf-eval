```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("./summary.csv")

data.head()
```

```python
fig,ax=plt.subplots(figsize=(25,10))
sns.barplot(data["University Name"],data.Performance, color="blue")
ax.set_title("Chart ranking for all the performance metric ",pad=20)
ax.set_xlabel("Names of Universities",labelpad=20)
ax.set_ylabel("Performance",labelpad=20)
plt.xticks(rotation=90)

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.1%}'.format(height), (x+0.25, y + height + 0.01))
    
plt.savefig('./data/perf-all.png')
```
